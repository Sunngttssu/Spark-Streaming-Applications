```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import time

# Replace with the path to your video file
video_file_path = "/content/WhatsApp Video 2024-11-13 at 18.41.15_0edd3fb3 (online-video-cutter.com).mp4"

# Open the video file
video_capture = cv2.VideoCapture(video_file_path)

if not video_capture.isOpened():
    print("Error opening video stream or file")
    exit()

frame_count = 0  # Initialize a frame counter

while True:
    # Read a frame from the video stream
    ret, frame = video_capture.read()

    # If frame is read correctly, ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame using cv2_imshow for Colab
    cv2_imshow(gray)

    # Extract the pixel values
    pixels = np.array(gray)
    print(f"Frame {frame_count} pixels:\n{pixels}\n")

    frame_count += 1
    if frame_count >= 10:  # Break the loop after processing 10 frames
        break

    time.sleep(0.01)

# When everything done, release the capture
video_capture.release()
```
